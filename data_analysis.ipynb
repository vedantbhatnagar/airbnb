```python
df_reg = df2[reg_columns]
df_reg.drop(columns = ['neighbourhood_group_cleansed', 'bathrooms', 'calendar_updated', 'license', 'latitude', 'longitude'], axis = 1, inplace = True)
```

```python
df_reg_nonull = df_reg.dropna()
```

```python
df_reg_nonull.shape
```

```python
X = df_reg_nonull.drop('price', axis = 1)
X
y = df_reg_nonull['price']
y
```

```python

# import the regressor
from sklearn.tree import DecisionTreeRegressor 
  
# create a regressor object
regressor = DecisionTreeRegressor(random_state = 0, max_depth = 6) 
  
# fit the regressor with X and Y data
regressor.fit(X, y)
```

```python
# import export_graphviz
from sklearn.tree import export_graphviz 
  
# export the decision tree to a tree.dot file
# for visualizing the plot easily anywhere
export_graphviz(regressor, out_file ='tree.gif', feature_names = X.columns) 
```

```python
df_reg_nonull['price_pred'] = regressor.predict(X)
```

```python
from sklearn.metrics import r2_score
r2_score(df_reg_nonull['price'], df_reg_nonull['price_pred'])
```
